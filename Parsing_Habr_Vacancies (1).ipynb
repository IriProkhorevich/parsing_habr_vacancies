### Алгоритм парсинга данных по вакансиям для IT-специалистов с одного из ресурсов: hh.ru, career.habr.com/vacancies и т.д.

-Алгоритм позволяет получить информацию о количестве вакансий по направлениям Аналитика данных и Data Science в разрезе уровней (Junior, Middle, Senior)

-За основу можно взять как один регион (НН, Москва и т.д.)  так и несколько.

-Результат представить в табличном и/или графическом виде.


```python
 # Импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup 
import time

pd.options.display.float_format = '{:.2f}'.format
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
page_number = 1
query = 'Аналитика данных'
region = 'c_678'   # 'c_678' Код региона Москва
data = []
while True:

  url = 'https://career.habr.com/vacancies?type=all' # # URL страницы с вакансиями career.habr.com/vacancies
  params = {
          'q': query,
          'page' : page_number ,
          'locations[]' : region
          }
  time.sleep(0.3)
  response = requests.get(url, params=params)   # Отправляем GET-запрос на страницу
  soup = BeautifulSoup(response.text, 'html.parser')  # Используем BeautifulSoup для парсинга HTML-кода страницы
  vacancies = soup.find_all('div', class_= 'vacancy-card__inner')
  for vacancy in vacancies:                           # Находим вакансии с заданными параметрами для поиска
    grade = vacancy.find('div', class_= 'vacancy-card__skills').text   # вытаскиваем грейд из каждой вакансии
    title = vacancy.find('div', class_ = 'vacancy-card__title').text   # вытаскиваем название каждой вакансии
    data.append([title,grade ])   # собираем все вакансии со всех страниц в список

  page_number += 1         # Увеличиваем номер страницы для перехода к следующей

  if not vacancies:  # прописываем условия завершения цикла while
     break
print(f'Всего вакансий - Аналитика данных:{len(data)}')  # выведем всего вакансий с заданными параметрами по поиску

```

```python
header = ['title' , 'grade' ]
df = pd.DataFrame(data, columns = header)  # запишем все вакансии в датафрейм
df
```

```python
# Соберем нужные нам грейды по вакансиям
df.loc[df['grade'].str.contains('Junior', case=False), 'Junior'] = 'Junior'
df.loc[df['grade'].str.contains('Middle', case=False), 'Middle'] = 'Middle'
df.loc[df['grade'].str.contains('Senior', case=False), 'Senior'] = 'Senior'
df.head()
```

```python
# Посчитаем и выведем в табличном виде информацию о количестве вакансий по направлениям Аналитика данных  в разрезе уровней (Junior, Middle, Senior).
df_analitics_vacancies = df.agg({'Middle': ['count'],
        'Junior' : ['count'],
        'Senior': ['count']
        })
df_analitics_vacancies
```
